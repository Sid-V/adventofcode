vertical math?

123 328  51 64 
 45 64  387 23 
  6 98  215 314
'*' '+'  '*' '+' 
  
123 * 45 * 6 = 33210 and so on

Grand total = 33210 + 490 + 4243455 + 401 = 4277556

```python
# read input
with open('input.txt', 'r') as file:
    lines = file.read().splitlines()

# Last line has operations and number of problems
operations = lines[-1].split()
num_problems = len(operations)

# Without the operations line
values = lines[:-1]
```

```python
# part 1
# Convert values to a grid of integers
grid = []
for line in values:
    row = [int(x) for x in line.split()]
    grid.append(row)

# Calculate every column
total = 0
for col_idx in range(num_problems):
    column_values = [grid[row_idx][col_idx] for row_idx in range(len(grid))]
    op = operations[col_idx]
    
    if op == '*':
        result = 1
        for val in column_values:
            result *= val
    elif op == '+':
        result = sum(column_values)
    
    total += result

print("Total:", total)
```

```python
# part 2

# reading is a bit different, reading right to left and top to bottom 
# 123 328  51 64 
#  45 64  387 23 
#   6 98  215 314
# '*' '+'  '*' '+' 

# padding line with spaces so we can transpose the grid
max_len = max(len(line) for line in lines)
lines = [line.ljust(max_len) for line in lines]
# transpose
grid = list(zip(*lines))

total = 0
# iterating left to right and then top to bottom
for col in grid:
    # get last character
    op_char = col[-1]
    
    # if its not a space, that means we can get our operand
    if op_char != ' ':
        op = op_char
        if op == '*':
            result = 1
        else:
            result = 0
        
    # get all the numbers in the column, excluding the last character which could be the operand    
    num = "".join(col[:-1]).strip()
    # perform operation
    if num:
        if op == '*':
            result *= int(num)
        else:
            result += int(num)

    # else, we are in between 2 different columns, we can add to result
    else:
        total += result

total += result
print("Total: ", total)
```
