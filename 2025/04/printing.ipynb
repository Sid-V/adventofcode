can access paper only if there are < 4 rolls around it

>=4 empty spaces

```python
# part 1
def get_cell(grid, i, j, rows, cols):
    if 0 <= i < rows and 0 <= j < cols:
        return grid[i][j]
    return 0  # Out of bounds = empty


with open('input.txt', 'r') as file:
    lines = file.read()

lines = lines.replace('.', '0').replace('@', '1')



grid = [[int(c) for c in line] for line in lines.strip().split('\n')]
rows = len(grid)
cols = len(grid[0])
```

```python
# part 1

ct = 0
for i in range(rows):
    for j in range(cols):
        cell = grid[i][j]
        if cell == 1:
            # check all 8 adjacent cells
            adjacent_cells = [
                get_cell(grid, i-1, j-1, rows, cols), get_cell(grid, i-1, j, rows, cols), get_cell(grid, i-1, j+1, rows, cols),
                get_cell(grid, i, j-1, rows, cols),                                        get_cell(grid, i, j+1, rows, cols),
                get_cell(grid, i+1, j-1, rows, cols), get_cell(grid, i+1, j, rows, cols), get_cell(grid, i+1, j+1, rows, cols)
            ]
            if sum(adjacent_cells) < 4:
                ct += 1

print(ct)
```

```python
# part 2

ct = 0
changed = True

# if we go through an entire pass without any changes then we are done

while changed:
    changed = False
    for i in range(rows):
        for j in range(cols):
            if grid[i][j] == 1:
                adjacent_cells = [
                    get_cell(grid, i-1, j-1, rows, cols), get_cell(grid, i-1, j, rows, cols), get_cell(grid, i-1, j+1, rows, cols),
                    get_cell(grid, i, j-1, rows, cols),                                        get_cell(grid, i, j+1, rows, cols),
                    get_cell(grid, i+1, j-1, rows, cols), get_cell(grid, i+1, j, rows, cols), get_cell(grid, i+1, j+1, rows, cols)
                ]
                if sum(adjacent_cells) < 4:
                    grid[i][j] = 0
                    ct += 1
                    changed = True

print(ct)
```
