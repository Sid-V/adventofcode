1st line, press 1st button to turn on switch 3


```python
from itertools import combinations

# read input
with open("input.txt", "r") as f:
    lines = f.read()

# for diagram, *buttons, joltage in map(lines.strip().split()):
#     diagram = [c == '#' for c in diagram[1:-1]]
#     buttons = [eval(b[:-1]+',)') for b in buttons]
#     joltage = eval(joltage[1:-1])
#     numbers = range(len(joltage))

part1 = 0

for line in lines.strip().split('\n'):
    parts = line.split()
    diagram, *buttons, joltage = parts
    diagram = [c == '#' for c in diagram[1:-1]]
    buttons = [eval(b[:-1]+',)') for b in buttons]
    joltage = eval(joltage[1:-1])
    numbers = range(len(joltage))

    def toggle(buttons):
        for n in numbers:
            for pressed in combinations(buttons, n):
                lights = []
                for i in numbers:
                    toggle_count = 0
                    for p in pressed:
                        if i in p:
                            toggle_count += 1
                    is_on = toggle_count % 2 == 1
                    lights.append(is_on)
                if lights == diagram:
                    return n
                
    part1 += toggle(buttons)    


print("part1:", part1)
```
