```python
# Read input
with open("input.txt", "r") as file:
    lines = file.read()

ranges, ids = [
    part.split("\n") for part in lines.split("\n\n")
]
ranges = [r.split("-") for r in ranges]
```

```python
# part 1

# go through every id and check if it's in any of the ranges
ct = 0
for id in ids:
    id = int(id)
    for r in ranges:
        start, end = map(int, r)
        if start <= id <= end:
            ct += 1
            break
    

print("Fresh count:", ct)
```

```python
# part 2
# given the ranges, find the list of all the id values that are in the ranges
# How many id values are in the ranges?

# Option 1
# build a set and len(set)
## tooo slow!!!

# values = set()
# for r in ranges:
#     start, end = map(int, r)
#     for i in range(start, end + 1):
#         values.add(i)
# print("total count:", len(values))

# Option 2
# merge the ranges and overlapping ranges

merged = []
sorted_ranges = sorted(ranges, key=lambda x: int(x[0])) # sort by 1st value

# add in 1st range to final merged output
merged = [[int(sorted_ranges[0][0]), int(sorted_ranges[0][1])]]

for r in sorted_ranges:
    start, end = int(r[0]), int(r[1])
    
    # get the last value currently in the merged list
    lastEnd = merged[-1][1]
    
    # extend
    if start <= lastEnd:
        merged[-1][1] = max(lastEnd, end)
    # add new range
    else:
        merged.append([start, end])

# calculate the total number of values in the merged ranges
total = sum(end - start + 1 for start, end in merged)

print("Total count of fresh IDs:", total)

                       
```
