21 splits

```python
# read input

with open("input.txt", "r") as file:
    lines = file.read().strip().split("\n")


grid = [list(line) for line in lines]

# print the row and col index around the grid
# for r in grid:
#     print(" ".join(r))
```

```python
# part 1
# count number of splits

# Find starting position 'S'
start_index = grid[0].index('S')

# which cols are the beams in?
beams  = {start_index}
splits = 0

# skip row 0
for row_idx in range(1, len(grid)):
    new_beams = set()

    for col in beams:
        if col < 0 or col >= len(grid[row_idx]):
            continue
        
        cell = grid[row_idx][col]
        if cell == '^':
            splits += 1
            grid[row_idx][col] = '^'
            # split the beams into left and right of '^'
            new_beams.add(col-1)
            new_beams.add(col+1)
        else:
            grid[row_idx][col] = '|'
            new_beams.add(col)
    
    # update beams
    beams = new_beams

# print the row and col index around the grid
# print("--------End grid--------")
# for r in grid:
#     print(" ".join(r))

print("Number of splits:", splits)
            
```

```python
# part 2

# count number of possible timelines of the grid
# the beam can take either a left or right, and at each ^ it forms a new timeline
# we need to count the number of possible timelines
# we keep checking for the same (row, col) multiple times in each timeline, so we can memoize the tuple
memo = {}

def count_timelines(row, col):

    # base case
    if row >= len(grid):
        return 1
    
    if (row, col) in memo:
        return memo[(row, col)]
    
    cell = grid[row][col]

    if cell == '^':
        left = count_timelines(row + 1, col - 1)
        right = count_timelines(row + 1, col + 1)
        result = left + right
    else:
        result = count_timelines(row + 1, col)

    memo[(row, col)] = result
    return result

# Find starting position 'S'
start_index = grid[0].index('S')

total_timelines = count_timelines(1, start_index)
print("Total number of possible timelines:", total_timelines)
```
