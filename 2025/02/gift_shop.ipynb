https://adventofcode.com/2025/day/2

find invalid id
leading zeroes is not an ID

It needs to repeat

1 digit
2 digit
3 digit
upto length // 2 needs to repeat


```python
# part 1

with open("input.txt", "r") as file:
    data = file.read()

ranges = data.split(",")
sum = 0

for r in ranges:
    start, end = map(int, r.split("-"))

    # print("start:", start, "end:", end)
    for num in range(start, end + 1):
        s = str(num)
        first_half = s[:len(s)//2]
        second_half = s[len(s)//2:]

        if first_half == second_half:
            # print("Found:", num)
            sum += num

print("Sum: ", sum)
        
```

```python
# part 2

with open("input.txt", "r") as file:
    data = file.read()

ranges = data.split(",")
sum = 0

for r in ranges:
    start, end = map(int, r.split("-"))
    for num in range(start, end + 1):
        s = str(num)
        # Check if the number is made up of repeated patterns
        # start with 2 upto len(s)
        # each time we increment part length by dividing len(s) by k

        for k in range(2, len(s) + 1):
            if len(s) % k == 0:
                potentiallyInvalid = True
                part_length = len(s) // k
                i = 0
                while i < len(s):
                    if s[i:i+part_length] != s[:part_length]:
                        potentiallyInvalid = False
                        
                    i += part_length

                # break after finding an invalid number so we don't double count
                if potentiallyInvalid:
                    # print("Found:", num)
                    sum += num
                    break

print("Sum: " , sum)
```
