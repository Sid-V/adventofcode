https://adventofcode.com/2025/day/1

0-99 dial

starts at 50

ans = no. of times the dial is left pointing at 0 after any rotation in the sequence


```python
with open("2025_day1_input.txt", "r") as f:
    lines = f.read().splitlines()

#print(lines)

pos = 50
at_zero = 0
crosses_zero = 0

for line in lines:
    dir = line[0]
    rotations = int(line[1:])
    print(dir, rotations)

    # We need to track how many times we cross 0
    old_pos = pos

    if dir == "R":
        pos += rotations
        # for cases where we are at exactly 100
        crosses_zero += (pos // 100) - (old_pos // 100)
        
    elif dir == "L":
        pos -= rotations
        crosses_zero += (old_pos // 100) - (pos // 100)
    

    print("crosses zero:", crosses_zero)
    pos %= 100
    
    print("new pos:", pos)

    if pos == 0:
        at_zero += 1
        crosses_zero -= 1 # don't double count

print("no. of times it was at pos 0:", at_zero)
print("no. of times it crossed zero:", crosses_zero)
print("total:", at_zero + crosses_zero)
```
