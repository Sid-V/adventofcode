https://adventofcode.com/2025/day/1

0-99 dial

starts at 50

ans = no. of times the dial is left pointing at 0 after any rotation in the sequence


```python
with open("2025_day1_input.txt", "r") as f:
    lines = f.read().splitlines()

pos = 50
at_zero = 0
crosses_zero = 0

for line in lines:
    dir = line[0]
    rotations = int(line[1:])
    # print(dir, rotations)
    # LMAO
    for _ in range(rotations):
        if dir == "R":
            pos = (pos + 1) % 100
        else :  # dir == "L"
            pos = (pos - 1 + 100) % 100
        
        if pos == 0:
            crosses_zero += 1
    # print("crosses zero:", crosses_zero)
    # print("new pos:", pos)

    if pos == 0:
        at_zero += 1

print("no. of times it was at pos 0:", at_zero)
print("no. of times it crossed zero:", crosses_zero)
```
