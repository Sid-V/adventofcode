How many assignments pairs does 1 range fully contain the other?

eg. 49-89,49-88 
49-88 is fully inside 49-89

```python
import os

with open("2022_day4_input.txt", "r") as f:
    lines = f.read()

count_overlap = 0
total_pairs = 0
count_partial_overlap = 0
for line in lines.split("\n"):
    if line == "":
        continue
    total_pairs += 1
    pair1, pair2 = line.split(",")
    start1, end1 = map(int, pair1.split("-"))
    start2, end2 = map(int, pair2.split("-"))

    # full overlap
    if (start1 <= start2 and end1 >= end2) or (start2 <= start1 and end2 >= end1):
        # print(f"Full overlap found: {line}")
        count_overlap += 1
    
    # partial overlap
    if (start1 <= end2 and end1 >= start2):
        # print(f"Partial overlap found: {line}")
        count_partial_overlap += 1

print(f"Total full overlaps: {count_overlap}")
print(f"Total partial overlaps: {count_partial_overlap}")
```
