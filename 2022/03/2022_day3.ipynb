Find the item type that appears in both compartments of each rucksack. What is the sum of the priorities of those item types?

Q. Are we guaranteed that there WILL be an item that is in both compartments?

```python
import os


priority = {
    "a" : 1,
    "b" : 2,
    "c" : 3,
    "d" : 4,
    "e" : 5,
    "f" : 6,
    "g" : 7,
    "h" : 8,
    "i" : 9,
    "j" : 10,
    "k" : 11,
    "l" : 12,
    "m" : 13,
    "n" : 14,
    "o" : 15,
    "p" : 16,
    "q" : 17,
    "r" : 18,
    "s" : 19,
    "t" : 20,
    "u" : 21,
    "v" : 22,
    "w" : 23,
    "x" : 24,
    "y" : 25,
    "z" : 26,
    "A" : 27,
    "B" : 28,
    "C" : 29,
    "D" : 30,
    "E" : 31,
    "F" : 32,
    "G" : 33,
    "H" : 34,
    "I" : 35,
    "J" : 36,
    "K" : 37,
    "L" : 38,
    "M" : 39,
    "N" : 40,
    "O" : 41,
    "P" : 42,
    "Q" : 43,
    "R" : 44,
    "S" : 45,
    "T" : 46,
    "U" : 47,
    "V" : 48,
    "W" : 49,
    "X" : 50,
    "Y" : 51,
    "Z" : 52,
}

with open("2022_day3_input.txt", "r") as f:
    lines = f.read()

sum_priorities = 0
for line in lines.split("\n"):
    if line.strip() == "":
        continue
    # create sets for each half
    first_half = set(line[:len(line)//2])
    second_half = set(line[len(line)//2:])

    common = first_half.intersection(second_half)
    for item in common:
        # print(f"Common item: {item} with priority {priority[item]}")
        sum_priorities += priority[item]

print(f"Sum of priorities: {sum_priorities}")
```

```python
# part 2

# every 3 lines is a group
sum_badge_priorities = 0
lines_list = lines.split("\n")

for i in range(0, len(lines_list), 3): # step size of 3
    if lines_list[i].strip() == "":
        continue
    first_elf = set(lines_list[i])
    second_elf = set(lines_list[i+1])
    third_elf = set(lines_list[i+2])

    common_badge = first_elf.intersection(second_elf).intersection(third_elf)
    # print(f"Group {i//3 + 1} common badge items: {common_badge}")
    # there should only be one common badge item
    
    for item in common_badge:
        # print(f"Common badge item: {item} with priority {priority[item]}")
        sum_badge_priorities += priority[item]
print(f"Sum of badge priorities: {sum_badge_priorities}")
```
