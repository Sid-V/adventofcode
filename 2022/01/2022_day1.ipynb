input 
calories


each elf in 1 line


```python
import os
with open("2022_day1_input.txt", "r") as f:
    lines = f.read()

elves = lines.strip().split("\n\n")

# print(elves)

elf_totals = []
for elf in elves:
    calories = [(int(x)) for x in elf.split("\n")]
    elf_totals.append(sum(calories))

print("max: ", max(elf_totals))
```

```python
# top 3 elves summed

elf_totals.sort(reverse=True)
print("top 3:", sum(elf_totals[:3]))
```
