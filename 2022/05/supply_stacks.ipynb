tower of hanoi style problem with stacks

input reading is complex so split it to 2 parts

[P]     [L]         [T]            
[L]     [M] [G]     [G]     [S]    
[M]     [Q] [W]     [H] [R] [G]    
[N]     [F] [M]     [D] [V] [R] [N]
[W]     [G] [Q] [P] [J] [F] [M] [C]
[V] [H] [B] [F] [H] [M] [B] [H] [B]
[B] [Q] [D] [T] [T] [B] [N] [L] [D]
[H] [M] [N] [Z] [M] [C] [M] [P] [P]
 1   2   3   4   5   6   7   8   9 

move 8 from 3 to 2
move 1 from 9 to 5
move 5 from 4 to 7
move 6 from 1 to 4
move 8 from 6 to 8

```python
import os

with open("input.txt", "r") as f:
    lines = f.read()

stacktext, instructions = [
    part.split("\n") for part in lines.split("\n\n")
]

count_stacks = int(stacktext[-1].split()[-1])

# create 9 stacks
stacks = [[] for _ in range(count_stacks)]

# add crates to stacks
for line in stacktext[:-1]:
    # print("line:", line)
    for i in range(count_stacks):
        crate = line[i * 4 + 1]
        # print("crate:", crate)
        if crate != " ":
            stacks[i].insert(0, crate)

```

```python
# part 1
# go to the instructions
for line in instructions:
    parts = line.split()
    count = int(parts[1]) # 2nd part
    from_stack = int(parts[3]) # 4th part
    to_stack = int(parts[5]) # 6th part
    # print("move", count, "from", from_stack, "to", to_stack)
    for _ in range(count):
        item = stacks[from_stack - 1].pop()
        stacks[to_stack - 1].append(item)

# top of the stacks
result = "".join([stack[-1] for stack in stacks])
print("result:", result)
```

```python
# part 2
# can pick up multiple crates at once and maintain those order

for line in instructions:
    parts = line.split()
    count = int(parts[1]) # 2nd part
    from_stack = int(parts[3]) # 4th part
    to_stack = int(parts[5]) # 6th part
    # print("move", count, "from", from_stack, "to", to_stack)
    items = stacks[from_stack - 1][-count:]  # get the last 'count' items
    
    # pythonic way to return all items but the "last count" items
    stacks[from_stack - 1] = stacks[from_stack - 1][:-count] 
    
    stacks[to_stack - 1].extend(items)  # add "count" number of items to the new stack

# top of the stacks
result = "".join([stack[-1] for stack in stacks])
print("result:", result)
```
