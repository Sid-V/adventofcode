https://adventofcode.com/2022/day/2

rock paper scissors

A - rock
B - paper
C - scissors

X - Rock
Y - Paper
Z - Scissors

Cannot win every time, too suspicious

Winner = player with highest score

Score for 1 round = 1 (rock) / 2 (paper) / 3 (scissors) + outcome of round (0 for lose, 3 for draw, 6 for win)

```python
import os 

with open("2022_day2_input.txt", "r") as f:
    lines = f.read()

input = []
output = []

shape_score = {
    "X": 1,  # Rock
    "Y": 2,  # Paper
    "Z": 3   # Scissors
}

outcome_score = {
    ("A", "X"): 3,  # Rock vs Rock = Draw
    ("A", "Y"): 6,  # Rock vs Paper = Win
    ("A", "Z"): 0,  # Rock vs Scissors = Lose
    ("B", "X"): 0,  # Paper vs Rock = Lose
    ("B", "Y"): 3,  # Paper vs Paper = Draw
    ("B", "Z"): 6,  # Paper vs Scissors = Win
    ("C", "X"): 6,  # Scissors vs Rock = Win
    ("C", "Y"): 0,  # Scissors vs Paper = Lose
    ("C", "Z"): 3   # Scissors vs Scissors = Draw
}

for line in lines.split("\n"):
    if line.strip() == "":
        continue
    a, b = line.split()
    input.append(a)
    output.append(b)

print(input)
print("-----")
print(output)

total_score = 0

for i in range(len(input)):
    a = input[i]
    b = output[i]
    total_score += shape_score[b] + outcome_score[(a, b)]

print(total_score)
```

```python
# part 2
desired_outcome_score = {
    "X": 0,  # Lose
    "Y": 3,  # Draw
    "Z": 6   # Win
}

shape_to_play = {
    ("A", "X"): 3,  # vs Rock, need to Lose -> play Scissors (3)
    ("A", "Y"): 1,  # vs Rock, need to Draw -> play Rock (1)
    ("A", "Z"): 2,  # vs Rock, need to Win -> play Paper (2)
    ("B", "X"): 1,  # vs Paper, need to Lose -> play Rock (1)
    ("B", "Y"): 2,  # vs Paper, need to Draw -> play Paper (2)
    ("B", "Z"): 3,  # vs Paper, need to Win -> play Scissors (3)
    ("C", "X"): 2,  # vs Scissors, need to Lose -> play Paper (2)
    ("C", "Y"): 3,  # vs Scissors, need to Draw -> play Scissors (3)
    ("C", "Z"): 1,  # vs Scissors, need to Win -> play Rock (1)
}

new_total_score = 0
for i in range(len(input)):
    a = input[i]
    b = output[i]
    new_total_score += desired_outcome_score[b] + shape_to_play[(a, b)]

print(new_total_score)
```
